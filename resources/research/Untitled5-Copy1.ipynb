```python
import pandas as pd 
from bloomberg import BDP, BDH, BQL
```

```python
universe = 'JALSH Index'

fields = [
"SECURITY_NAME",
"INDUSTRY_SECTOR",
"INDUSTRY_GROUP",
"GICS_SUB_INDUSTRY_NAME"
]
```

```python
field_string = ', '.join(fields)
query = f"get({field_string})for(members(['{universe}']))"
```

```python
query
```

```python
results = BQL(query)
```

```python
results
```

```python
results['field'] = results['field'].str.replace("(fa_period_reference=range(2000,2024),fa_period_type='BA')", "")
results['value'] = pd.to_numeric(results['value'].replace("NaN", float("NaN")))
results['secondary_value'] = pd.to_datetime(results['secondary_value'])
```

```python
results.to_excel(f"{universe}_dataset_2000_2024.xlsx")
```

```python
data = results.pivot_table(values="value", index="secondary_value", columns=["security", "field"])
data = data.resample("YS").last()
data.index.name = None
data.columns.names = [None, None]
```

```python
data.to_csv(f"{universe}_dataset_2000_2024_clean.csv",)
```
